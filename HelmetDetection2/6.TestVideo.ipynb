```python
from Module.ImagePrediction import GetPrediction
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
TestVideoPos = r'E:\GithubProjects\KagglePractice\HelmetDetection2\TestVideo.mp4'
```

```python
cap = cv.VideoCapture(TestVideoPos)

fig = plt.figure()
imgs = []

i = 0
while True:
	ret, img = cap.read()
	im = GetPrediction(img)
	imgs.append([im])

	# if cv.waitKey() == ord('q'):
	# 	break
	i += 1
	print(i)
	if i == 2:
		break
cap.release()

ani = animation.ArtistAnimation(
	fig, imgs, interval = 50, blit = True, repeat_delay = 1000
)
ani.save(r'E:\GithubProjects\KagglePractice\HelmetDetection2\record.gif')
```
