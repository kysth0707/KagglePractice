```python
import cv2 as cv
import copy
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def Rotate(ResultImage, angle):
	ResultImage = copy.deepcopy(ResultImage)
	image_center = tuple(np.array(ResultImage.shape[1::-1]) / 2)
	rot_mat = cv.getRotationMatrix2D(image_center, angle, 1.0)
	ResultImage = cv.warpAffine(ResultImage, rot_mat, ResultImage.shape[1::-1], flags=cv.INTER_LINEAR)
	return ResultImage
```

```python
TargetPos = r'E:\GithubProjects\KagglePractice\HelmetDetection2\Datasets\original_helmet'
SavePos = r'E:\GithubProjects\KagglePractice\HelmetDetection2\Datasets\helmet'
ImageList = np.array([cv.imread(f"{TargetPos}/{file}") for file in os.listdir(TargetPos)])
```

```python
plt.imshow(cv.cvtColor(ImageList[0], cv.COLOR_BGR2RGB))
```

```python
RotateList = [i+1 for i in range(20)] + [i-20 for i in range(20)]
```

```python
OutputImage = []

for Image in ImageList:
	for RotateAngle in RotateList:
		OutputImage.append(Rotate(Image, RotateAngle))
```

```python
Index = 0
for Image in OutputImage:
	cv.imwrite(fr"{SavePos}\{Index}.png", Image)
	Index += 1
	print(f"\r{Index} / {len(OutputImage)}",end="")
```
