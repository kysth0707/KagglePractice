```python
import shutil
import os
import random
import copy
```

```python
TrainPos = r'E:\GithubProjects\KagglePractice\HelmetDetection\Datasets\train'
ValPos = r'E:\GithubProjects\KagglePractice\HelmetDetection\Datasets\val'
```

```python
shutil.rmtree(TrainPos)
shutil.rmtree(ValPos)
```

```python
os.mkdir(TrainPos)
os.mkdir(fr'{TrainPos}\with_helmet')
os.mkdir(fr'{TrainPos}\without_helmet')
os.mkdir(ValPos)
os.mkdir(fr'{ValPos}\with_helmet')
os.mkdir(fr'{ValPos}\without_helmet')
```

```python
HelmetImages = r'E:\GithubProjects\KagglePractice\HelmetDetection\Datasets\clearedwithhelmet'
WithoutHelmetImages = r'E:\GithubProjects\KagglePractice\HelmetDetection\Datasets\clearedwithouthelmet'
```

```python
HelmetImageCount = len(os.listdir(HelmetImages))
WithoutHelmetImageCount = len(os.listdir(WithoutHelmetImages))
HelmetImageCount, WithoutHelmetImageCount
```

```python
HelmetImageList = [i for i in range(HelmetImageCount)]
WithoutHelmetImageList = [i for i in range(WithoutHelmetImageCount)]
# HelmetImageList, WithoutHelmetImageList
```

```python
ChooseCount = 100
```

```python
ChooseNumHelmet = []
ChooseNumWithoutHelmet = []
for _ in range(ChooseCount):
	RamdomInt = random.randint(1, len(HelmetImageList))
	ChooseNumHelmet.append(copy.deepcopy(HelmetImageList[RamdomInt - 1]))
	del HelmetImageList[RamdomInt - 1]

	RamdomInt = random.randint(1, len(WithoutHelmetImageList))
	ChooseNumWithoutHelmet.append(copy.deepcopy(WithoutHelmetImageList[RamdomInt - 1]))
	del WithoutHelmetImageList[RamdomInt - 1]
# ChooseNumHelmet
```

```python
HelmetList = os.listdir(HelmetImages)
print(len(HelmetList))
for Num in ChooseNumHelmet:
	print(Num)
	TargetPos = fr"{HelmetImages}\{HelmetList[Num]}"
	SavePos = fr"{TrainPos}\with_helmet\{HelmetList[Num]}"
	shutil.move(TargetPos , SavePos)
for Num in HelmetImageList:
	TargetPos = fr"{HelmetImages}\{HelmetList[Num]}"
	SavePos = fr"{ValPos}\with_helmet\{HelmetList[Num]}"
	shutil.move(TargetPos , SavePos)
```

```python
WithoutHelmetList = os.listdir(WithoutHelmetImages)
print(len(WithoutHelmetList))
for Num in ChooseNumWithoutHelmet:
	print(Num)
	TargetPos = fr"{WithoutHelmetImages}\{WithoutHelmetList[Num]}"
	SavePos = fr"{TrainPos}\without_helmet\{WithoutHelmetList[Num]}"
	shutil.move(TargetPos , SavePos)
for Num in WithoutHelmetImageList:
	TargetPos = fr"{WithoutHelmetImages}\{WithoutHelmetList[Num]}"
	SavePos = fr"{ValPos}\without_helmet\{WithoutHelmetList[Num]}"
	shutil.move(TargetPos , SavePos)
```
